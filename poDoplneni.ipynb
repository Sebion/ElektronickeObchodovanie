```python
hlavna_sorted = hlavna.sort_values('AucZac')
hlavna_sorted.head(50)
```

```python
columns_to_drop = ['PocetPrihlasenychUcast', 'PocKolERMMA', 'PocKolZadavac', 'PocKolNIPPON', 'PocKolHOLLAND', 'UsporaPorov', 'UsporaVstup', 'UsporaAukc', 'KoefVarPonukCelk', 'KoefVarPonukERMNIP']
hlavna_sorted = hlavna_sorted.drop(columns_to_drop, axis=1)
```

```python
hlavna_sorted.head()
```

```python
grouped = hlavna_sorted.groupby('Org')
sorted_grouped = grouped.apply(lambda x: x.sort_values(['Org', 'AucZac']))
sorted_grouped.head(500)
```

```python
# Group by 'Org' and 'Vitaz' and sort by 'AucZac'
hlavna_sorted = hlavna_sorted.sort_values(['Org', 'Vitaz', 'AucZac'])
# grouped = hlavna_sorted.groupby(['Org', 'Vitaz'])
grouped = hlavna_sorted.groupby(['Org', 'Subkateg'])


# Calculate cumulative count
hlavna_sorted['HistoryDealsInCategory'] = grouped.cumcount() + 1
```

```python
# Group by 'Org' and 'Vitaz' and sort by 'AucZac'
hlavna_sorted = hlavna_sorted.sort_values(['Org', 'Vitaz', 'AucZac'])
grouped = hlavna_sorted.groupby(['Org', 'Vitaz'])
# grouped = hlavna_sorted.groupby(['Org', 'Subkateg'])

# Calculate cumulative count
hlavna_sorted['HistoryDealsWithVitaz'] = grouped.cumcount() + 1
```

```python
# Group by 'Org' and 'Vitaz' and sort by 'AucZac'
hlavna_sorted = hlavna_sorted.sort_values(['Org', 'Vitaz', 'AucZac'])
# grouped = hlavna_sorted.groupby(['Org', 'Vitaz'])
grouped = hlavna_sorted.groupby(['Org', 'Subkateg'])


# Calculate cumulative sum
hlavna_sorted['MoneySpentSoFarInCategory'] = grouped['VitaznaPonuka'].cumsum()
```

```python
# Group by 'Org' and 'Vitaz' and sort by 'AucZac'
hlavna_sorted = hlavna_sorted.sort_values(['Org', 'Vitaz', 'AucZac'])
grouped = hlavna_sorted.groupby(['Org', 'Vitaz'])
# grouped = hlavna_sorted.groupby(['Org', 'Subkateg'])


# Calculate cumulative sum
hlavna_sorted['MoneySpentSoFarWithVitaz'] = grouped['VitaznaPonuka'].cumsum()
```

```python
hlavna_sorted = hlavna_sorted.sort_values(['Org', 'Subkateg', 'Vitaz', 'AucZac'])

hlavna_sorted.head(50)
```
